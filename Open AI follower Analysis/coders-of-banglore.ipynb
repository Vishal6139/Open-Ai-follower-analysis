```python
with open("finaldata.txt", encoding='utf-8') as f:
    data = f.read()
```

```python
chunks = data.split("\n\n") 
chunks = [c for c in chunks if len(c)>3]
```

```python
def parse_chunk(chunk): 
        chunk = chunk.strip()
        sep_chunk = chunk.split('\n')
        username = sep_chunk[0]
        no_of_posts = int(sep_chunk[1].split(" post")[0].replace(",", ""))
        no_of_followers = float(sep_chunk[2].split(" follower")[0].replace(",", "").replace("K", "").replace("M", ""))
        if("K" in sep_chunk[2]):
            no_of_followers = int(no_of_followers * 1000)
        elif("M" in sep_chunk[2]):
            no_of_followers = int(no_of_followers * 1000000)
        else:
            no_of_followers = int(no_of_followers)
             

    
        no_of_following = float(sep_chunk[3].split(" following")[0].replace(",", "").replace("K", "").replace("M", ""))
        if("K" in sep_chunk[3]):
            no_of_following = int(no_of_following * 1000)
        elif("M" in sep_chunk[3]):
            no_of_following = int(no_of_following * 1000000)
        else:
            no_of_following = int(no_of_following)

            
        name = sep_chunk[4]
        if(len(sep_chunk)> 5):
            type_of_page = sep_chunk[5]
            bio = "\n".join(sep_chunk[6:])
        else:
            type_of_page = "Unknown"
            bio = "" 
        return {"username": username, "no_of_posts": no_of_posts, "no_of_followers": no_of_followers, "type_of_page": type_of_page,
                "no_of_following": no_of_following, "name": name, "type_of_page": type_of_page, "bio": bio}
    
```

```python
all_chunks = [] 
for chunk in chunks:  
    parsed_chunk = parse_chunk(chunk)
    all_chunks.append(parsed_chunk) 
```

```python
import json
s = json.dumps(all_chunks, indent=4)  
with open("data.json", "w") as f:
    f.write(s)
```

```python
# who has the maximum posts
max = 0
for chunk in all_chunks:
    if(max<chunk['no_of_posts']):
        max = chunk['no_of_posts']
        chunk_with_max_post = chunk
print(chunk_with_max_post['username'])
        
```

```python
# who follows maximum people?
max = 0
for chunk in all_chunks:
    if(max<chunk['no_of_following']):
        max = chunk['no_of_following']
        chunk_with_max_following = chunk
print(chunk_with_max_following['username'])
```

```python
from pptx import Presentation
from pptx.util import Inches
from pptx.chart.data import ChartData
from pptx.enum.chart import XL_CHART_TYPE
import json

# Load your structured JSON data
with open("data.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Create presentation
prs = Presentation()

def add_chart_slide(title, categories, values):
    slide_layout = prs.slide_layouts[5]
    slide = prs.slides.add_slide(slide_layout)
    
    slide.shapes.title.text = title
    
    chart_data = ChartData()
    chart_data.categories = categories
    chart_data.add_series("Value", values)
    
    x, y, cx, cy = Inches(1), Inches(1.5), Inches(8), Inches(4.5)
    chart = slide.shapes.add_chart(
        XL_CHART_TYPE.COLUMN_CLUSTERED, x, y, cx, cy, chart_data
    ).chart
    
    chart.has_legend = False
    chart.value_axis.has_major_gridlines = True

# -------- Slide 1: Top 10 by Posts --------
sorted_posts = sorted(data, key=lambda x: x['no_of_posts'], reverse=True)[:10]

add_chart_slide(
    "Top 10 Users by Number of Posts",
    [u['username'] for u in sorted_posts],
    [u['no_of_posts'] for u in sorted_posts]
)

# -------- Slide 2: Top 10 by Followers --------
sorted_followers = sorted(data, key=lambda x: x['no_of_followers'], reverse=True)[:10]

add_chart_slide(
    "Top 10 Users by Number of Followers",
    [u['username'] for u in sorted_followers],
    [u['no_of_followers'] for u in sorted_followers]
)

# -------- Slide 3: Top 10 by Following --------
sorted_following = sorted(data, key=lambda x: x['no_of_following'], reverse=True)[:10]

add_chart_slide(
    "Top 10 Users by Number of Following",
    [u['username'] for u in sorted_following],
    [u['no_of_following'] for u in sorted_following]
)

# -------- Slide 4: Category Distribution --------
category_count = {}
for user in data:
    category = user['type_of_page']
    category_count[category] = category_count.get(category, 0) + 1

sorted_categories = sorted(category_count.items(), key=lambda x: x[1], reverse=True)

add_chart_slide(
    "Distribution of Page Categories",
    [c[0] for c in sorted_categories],
    [c[1] for c in sorted_categories]
)

# Save dashboard
prs.save("Instagram_PowerBI_Style_Dashboard.pptx")

print("Dashboard Created Successfully!")
```
